```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
csv_path = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(csv_path)

purchase_data

```

```python
# determine unique count of players

unique_players = len(purchase_data["SN"].value_counts())

unique_players

# table of players

total_players = pd.DataFrame({"Total Players" : [unique_players]})

total_players 
```

```python
# purchasing analysis (totals)

# set variables and formulas

unique_items = len(purchase_data["Item Name"].value_counts())

average_price = purchase_data["Price"].mean()

number_of_purchases = purchase_data["Purchase ID"].count()

total_revenue = purchase_data["Price"].sum()

# create summary DF and table to display

summary_purchasing_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                                          "Average Purchase Price": [average_price],
                                            "Number of Purchases": [number_of_purchases],
                                            "Total Revenue": [total_revenue]})

summary_purchasing_analysis

```

```python
# gender demographics

# groupby gender and sn

gender_count = purchase_data.groupby("Gender") ["SN"].nunique()

gender_count

gender_percent = gender_count / unique_players

gender_percent

gender_demo = pd.DataFrame({"Total Count" : gender_count,
                            "Percentage of Players": gender_percent})

gender_demo
```

```python
# purchasing analysis by gender

# groupyby gender

purchase_gender_group = purchase_data.groupby(["Gender"])

# set variables and formulas

purchase_count = purchase_gender_group["Purchase ID"].count()

purchase_count

avg_purchase_price = purchase_gender_group["Price"].mean()

avg_purchase_price

purchase_value = purchase_gender_group["Price"].sum()

purchase_value

avg_total_per = purchase_value / gender_count

avg_total_per

#create summary dataframe and display

purchase_analysis_gender = pd.DataFrame({"Purchase Count": purchase_count,
                                          "Average Purchase Price": avg_purchase_price,
                                            "Total Purchase Value": purchase_value,
                                            "Average Total Purchase per Person": avg_total_per})

purchase_analysis_gender
```

```python
# age demographics

# establish bins for ages

bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# add age group to dataset

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

purchase_data

# groupby the age group

age_group = purchase_data.groupby(["Age Group"])

age_group.count()

# calculations

age_total = age_group["SN"].nunique()

age_total

age_percent = age_total / unique_players

age_percent

# create summary dataframe and display 

age_demographics = pd.DataFrame({"Total Count": age_total,
                                "Percentage of Players": age_percent})

age_demographics
```

```python
# purchasing analysis by age group

# groupyby age group

purchase_age_group = purchase_data.groupby(["Age Group"])

# set variables and formulas

purchase_count_age = purchase_age_group["Purchase ID"].count()

purchase_count_age

avg_purchase_price_age = purchase_age_group["Price"].mean()

avg_purchase_price_age

purchase_value_age = purchase_age_group["Price"].sum()

purchase_value_age

avg_total_per_age = purchase_value_age / age_total

avg_total_per_age

#create summary dataframe and display

purchase_analysis_age = pd.DataFrame({"Purchase Count": purchase_count_age,
                                          "Average Purchase Price": avg_purchase_price_age,
                                            "Total Purchase Value": purchase_value_age,
                                            "Average Total Purchase per Person": avg_total_per_age})

purchase_analysis_age
```

```python
# top spenders

# groupby SN

sn_group = purchase_data.groupby(["SN"])

sn_group.count()

# set variables and formulas

purchase_count_sn = sn_group["Purchase ID"].count()

purchase_count_sn

avg_purchase_price_sn = sn_group["Price"].mean()

avg_purchase_price_sn

purchase_value_sn = sn_group["Price"].sum()

purchase_value_sn

# create summary dataframe

top_spenders = pd.DataFrame({"Purchase Count": purchase_count_sn,
                            "Average Purchase Price": avg_purchase_price_sn,
                            "Total Purchase Value": purchase_value_sn})

top_spenders

# sort by total purchase value

top_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)

top_spenders

```

```python
# most popular items

# groupby item id and item name

item_group = purchase_data.groupby(["Item ID", "Item Name"])

item_group.count()

# set variables and formulas

purchase_count_item = item_group["Purchase ID"].count()

purchase_count_item

item_price = item_group["Price"].sum()

item_price

purchase_value_item = purchase_count_item * item_price

purchase_value_item

# create summary dataframe

most_popular = pd.DataFrame({"Purchase Count": purchase_count_item,
                            "Item Price": item_price,
                            "Total Purchase Value": purchase_value_item})

most_popular

# sort by purchase count

most_popular = most_popular.sort_values("Purchase Count", ascending=False)

most_popular
```
